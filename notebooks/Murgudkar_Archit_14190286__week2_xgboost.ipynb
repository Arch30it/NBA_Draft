Deleting unwanted features:
1. "num" consists of player's number. Its better to delete it as this feature can cause our model to adapt specific patterns.
2. "type" consists of all same values, which wont help model in decision making.
3. "ht" consists of misleading information.
4. "pick" and "Rec_Rank" as it contains a lot of null values and imputing any value in the missing fields would generate bias.
5. "player_id" as it is a player specific information and this wont help in classifying unseen data.

```python
df_cleaned.drop(['num', 'type', 'ht', 'pick', 'Rec_Rank', 'player_id'], axis=1, inplace=True)
```

```python
df_test_cleaned.drop(['num', 'type', 'ht', 'pick', 'Rec_Rank', 'player_id'], axis=1, inplace=True)
```

```python
df_cleaned['yr'].fillna("Jr", inplace=True)
df_cleaned['yr'].replace(['None', '0','57.1','42.9'],'Jr', inplace=True)
df_cleaned['yr'].value_counts()
```

```python
df_test_cleaned['yr'].fillna("Jr", inplace=True)
df_test_cleaned['yr'].replace(['None', '0','57.1','42.9'],'Jr', inplace=True)
df_test_cleaned['yr'].value_counts()
```

```python
null_cols=df_cleaned.columns[df_cleaned.isna().any()]
null_cols
```

All above columns are in numeric format, its wise to impute the missing values with the median of each feature, as median doesn't gets affected because of any extreme outliers if present.

```python
for x in null_cols:
  df_cleaned.fillna(df_cleaned[x].median(),inplace=True)
```

```python
for x in null_cols:
  df_test_cleaned.fillna(df_test_cleaned[x].median(),inplace=True)
```

Implementing ordinal encoding on the "yr" feature:

```python
from sklearn.preprocessing import OrdinalEncoder
#initialise encoder:
encoder = OrdinalEncoder(
    categories=[['Fr', 'Jr', 'So', 'Sr']],
    handle_unknown="use_encoded_value",
    unknown_value=-1,
)

df_cleaned["yr_encoded"] = encoder.fit_transform(df_cleaned[["yr"]])
df_test_cleaned["yr_encoded"] = encoder.transform(df_test_cleaned[["yr"]])


#retrieve categories:
encoder.categories
```

```python
encoder = OrdinalEncoder(
    categories=[['Fr', 'Jr', 'So', 'Sr']],
    handle_unknown="use_encoded_value",
    unknown_value=-1,
)

df_test_cleaned["yr_encoded"] = encoder.fit_transform(df_test_cleaned[["yr"]])

#retrieve categories:
encoder.categories
```
